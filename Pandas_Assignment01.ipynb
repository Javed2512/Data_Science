```python
Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

Ans)In Pandas, a DataFrame is a two-dimensional tabular data structure, similar to a spreadsheet or a SQL table. It consists
of rows and columns, where each column can contain a different data type (e.g., integer, float, string, etc.). In essence,
a DataFrame is a collection of Pandas Series that share the same index.

On the other hand, a Pandas Series is a one-dimensional labeled array capable of holding data of any type 
(e.g., integer, float, string, etc.). A Series has a single column of data and an associated index, which labels the elements 
of the Series.
```

```python
import pandas as pd

# Create a Series
s = pd.Series([1, 2, 3, 4])

# Create a DataFrame
df = pd.DataFrame({'A': [1, 2, 3, 4], 'B': ['a', 'b', 'c', 'd']})

# Print the Series and DataFrame
print('Series')
print(s)
print('DataFrame')
print(df)
```

```python
Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can
you give an example of when you might use one of these functions?

Ans)There are numerous functions available in Pandas to manipulate data in a DataFrame. Here are some common functions:

head() and tail(): These functions are used to display the first n or last n rows of a DataFrame, respectively. For example,
df.head() will display the first 5 rows of the DataFrame df.

describe(): This function provides summary statistics of the DataFrame, including count, mean, standard deviation, minimum
and maximum values. For example, df.describe() will show summary statistics for all numerical columns in the DataFrame df.
```
